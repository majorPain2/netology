```python
#По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония)
df_power = pd.read_csv('power.csv')
df_power2 = df_power[(df_power['country'] == 'Latvia')|(df_power['country'] == 'Lithuania')|(df_power['country'] == 'Estonia')]
df_power3 = df_power2[(df_power2['category'] == 4)|(df_power2['category'] == 12)|(df_power2['category'] == 21)]
df_power3[(df_power3['quantity'] >= 0)&(df_power3['year'] >= 2005)&(df_power3['year'] <= 2010)].sum()['quantity']
```
