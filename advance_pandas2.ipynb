```python
import pandas as pd
rating = pd.read_csv('ml-latest-small\\ratings.csv')
rating
```

```python
agg_rating = rating.groupby('userId').agg({'timestamp': ['max','min'], 'movieId': 'count'})
agg_rating['difference'] = agg_rating['timestamp']['max'] - agg_rating['timestamp']['min']
int(agg_rating[agg_rating['movieId']['count'] > 100].reset_index().mean()['difference']/60/60/24)
#время жизни в днях
```
